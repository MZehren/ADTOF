```python
"""
go through all files recursively
"""
rootFodler = "../../ADTSets/B L O O D L I N E/raw"
rarFiles = []
zipFiles = []
unknownFiles = []
for root, dirs, files in os.walk(rootFodler):
    path = root.split(os.sep)
    print((len(path) - 1) * '---', os.path.basename(root))
    
    for file in files:
        print(len(path) * '---', file)
        fullPath =  os.sep.join(path + [file])
        
        if file[-3:] == "zip":
            zipFiles.append(fullPath)
        elif file[-3:] == "rar":
            rarFiles.append(fullPath)
        else:
            unknownFiles.append(fullPath)
```

```python
"""
decompress everything in a uncompressed folder
"""
from pyunpack import Archive

uncompressedFolder = "../../ADTSets/B L O O D L I N E/uncompressed"

for file in rarFiles:
    folderName =  uncompressedFolder + os.sep + file.split(os.sep)[-1][:-4]
    Archive(file).extractall(folderName, auto_create_dir="True")
    print(folderName)
    
for file in zipFiles:
    folderName =  uncompressedFolder + os.sep + file.split(os.sep)[-1][:-4]
    Archive(file).extractall(folderName, auto_create_dir="True")
    print(folderName)
```

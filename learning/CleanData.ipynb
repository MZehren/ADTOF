```python
"""
2 - Go through all files recursively and decompress them
"""
import os
# from pyunpack import Archive

rootFodler = "/home/mickael/Documents/Datasets/drumsTranscription/ADTOF/sodamlazy"
compressedFiles = []
unknownFiles = []

for root, dirs, files in os.walk(rootFodler):
    path = root.split(os.sep)
    
    for file in files:
        print(len(path) * '---', file)
        fullPath =  os.sep.join(path + [file])
        
        if file[-3:] == "zip" or file[-3:] == "rar":
            compressedFiles.append(fullPath)
        else:
            unknownFiles.append(fullPath)

for file in compressedFiles:
    folderName =  file[:-4]
    Archive(file).extractall(folderName, auto_create_dir="True")
    os.remove(file)
    print(folderName)
```

```python
"""
3 - Go throught all the files to see which tracks and which format are used
"""

def detectPhaseShift(files):
    return ("song.ogg" in files or "guitar.ogg" in files) and "song.ini" in files and "notes.mid" in files

def detectRockBand(File):
    #Check if the extension is known
    extensions= [".rba"]
    if any([True for i in extensions if i in file]):
        return True
    
    #Check if the file starts with the good string
    beginningOfFile = ["CON", "RBSF"]
    try:
        with open(file, "r") as f:
            firstLine = f.readline()
            if [True for i in beginningOfFile if beginningOfFile[:4] in firstLine].any():
                return True
    except:
        return False
    
phaseShiftTracks = []
rockBandTracks = []
for root, dirs, files in os.walk(rootFodler):
    if detectPhaseShift(files):
        phaseShiftTracks.append(root)
    else:
        for file in files:
            if detectRockBand(file):
                rockBandTracks.append(root)
        

print("\n".join(phaseShiftTracks))
print(rockBandTracks)
         
```
